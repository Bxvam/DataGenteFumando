```python
!rm -rf /content/smokingYolo

!git clone -b master https://github.com/hfuuliuyi/smokingYolo.git

import cv2
import glob

carpeta_imagenes = '/content/smokingYolo/imagesdata'
nombre_video = '/content/dataset_final.mp4'
fps = 30

imagenes = sorted(glob.glob(f"{carpeta_imagenes}/*.jpg") + glob.glob(f"{carpeta_imagenes}/*.png"))

if not imagenes:
    print(f"No se encontraron imágenes en: {carpeta_imagenes}.")
else:
    frame_base = cv2.imread(imagenes[0])
    alto, ancho, capas = frame_base.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    video = cv2.VideoWriter(nombre_video, fourcc, fps, (ancho, alto))

    print(f"Procesando {len(imagenes)} imágenes")
    for ruta in imagenes:
        img = cv2.imread(ruta)
        if img is not None:
            img_redimensionada = cv2.resize(img, (ancho, alto))
            video.write(img_redimensionada)

    video.release()
    print(f"Terminado, archivo {nombre_video} está listo")
```

```python
import cv2
import glob
import os
import re

carpeta_imagenes = '/content/smokingYolo/imagesdata'
nombre_video = '/content/dataset_definitivo.mp4'
fps = 30

imagenes_crudas = glob.glob(f"{carpeta_imagenes}/*.jpg") + glob.glob(f"{carpeta_imagenes}/*.png")

def ordenar_por_escena_y_numero(ruta):
    nombre_archivo = os.path.basename(ruta)
    match = re.match(r"(\d+)(.*)", nombre_archivo)
    if match:
        numero = int(match.group(1))
        texto_del_video = match.group(2)
        return (texto_del_video, numero)
    return (nombre_archivo, 0)

imagenes = sorted(imagenes_crudas, key=ordenar_por_escena_y_numero)

if not imagenes:
    print("No se encontraron imágenes")
else:
    frame_base = cv2.imread(imagenes[0])
    alto, ancho, capas = frame_base.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    video = cv2.VideoWriter(nombre_video, fourcc, fps, (ancho, alto))

    print(f"Procesando {len(imagenes)} imágenes agrupadas correctamente por escena")
    for ruta in imagenes:
        img = cv2.imread(ruta)
        if img is not None:
            img_redimensionada = cv2.resize(img, (ancho, alto))
            video.write(img_redimensionada)

    video.release()
    print(f"Terminado archivo {nombre_video} está listo")
```
